# Project 1: Create a Program in Python to Generate the Fibonacci Sequence

```
def fib(m):
    a, b = 0, 1
    for i in range(m):
        print(a)
        a, b = b, a+b
```

```
fib(7)
```

```
def fi(n):
    a, b = 0, 1
    L =[0]
    for i in range(n-1):
        a, b = b, a+b
        L.append(a)
    return L
```

```
fi(7)
```
