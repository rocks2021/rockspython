# Python For Else
* The code in the `else` block runs only if the loop completes without encountering a break statement.

```
# Print a message when the loop has ended.
for x in range(7):
    print(x)
else:
    print('Finally!')
```

```
for n in range(17, 27):
    for x in range(2, n):
        if n % x == 0:
            print( n, '=', x, '*', n//x)
            break
    else:        
        print(n, 'is a prime number.')
```

```
for num in range(11, 17):
    if num % 2 == 0:
        print(num, 'is an even number.')
        continue
    print(num, 'is an odd number.')
```

```

num = int(input('Enter an integer number: '))
if num > 1:
    for i in range(2, num):
        if (num % i) == 0:
            print(num, 'is not is a prime number.')
            break
    else:
        print(num, 'is a prime number.')

else:
    print(num, 'is not a prime number.')
```
