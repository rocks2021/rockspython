# Creat a Program in Python to Check if an Integer Is a Prime Number or Not

```
num = int(input('Enter an integer number: '))

if num > 1:
    for i in range(2, num):
        if (num % i) == 0:
            print(num, 'is not a prime number.')
            break
    else:
        print(num, 'is a prime number.')
else:
    print(num, 'is not a prime number.')
```
