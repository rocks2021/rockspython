# Lists and List Manipulations
## A list is a sequence.
* A list is a sequence, an ordered collection of items.
* You can access the items one at a time with the bracket operator.


```
numbers = [17, 27, 37, 47] # a list of four integers
type(numbers)
```

```
rocks = ['basalt', 'Harburgite', 'breccia', 'gabbro', 'sandstone'] # a list of four strings
type(rocks)
```

```
harzburgite = [209, 1207, 'igneous rock', [20,70]] # a nested list
type(harzburgite)
```

## List Slices
* The operator [n:m] returns the part of the list from the n-eth element to the m-eth element, including the first but excluding the last.
* The slice starts at the beginning if omitting the first index before the colon.
* The slice goes to the end if omitting the second index.
* The slice is a copy of the whole list if ommitting both.
* A slice operator on the left side of an assignment can update elements.
* If an index has a negative value, it counts backward from the end.

```
rocks = ['basalt', 'Harzburgite', 'breccia', 'gabbro', 'sandstone']
rocks[1:3]
```

```
rocks[:4]
```

```
rocks[3:]
```

```
rocks[:]
```

```
rocks[1:3] = ['mudstone', 'gneiss']
rocks
```

```
rocks = ['basalt', 'mudstone', 'gabbro', 'sandstone']
print(rocks[-1]) # Negative indices count from the end of the list.
print(rocks[-3: -1])
```

```
n = list(range(7)) # range creates a list of integers.
print(n)
print(n[2:4]) # a slice from index 2 to 4 (exlusive)
print(n[2: ]) # a slice from index 2 to the end.
print(n[ :2]) # a slice from the start to index 2 (exclusive)
print(n[:]) # a slice of the whole list
print(n[:-1]) # Negative indices count from the end of the list.
n[2:4] = [7,8]
print(n)
```

## List Methods
* append adds a new element to the end of the a list.
* extend takes a list as an argument and appends all of the elements.
* sort arranges the elements from low to high.

```
igneous = ['basalt', 'gabbro', 'pumice']
igneous.append('granite')
igneous
```

```
igneous1 = ['basalt', 'gabbro','pumice'] # igneous2 unmodified
igneous2 = ['granite', 'diabase']
igneous1.extend(igneous2)
igneous1
```

```
sedimentary = ['sandstone', 'mudstone', 'conglomerate', 'limestone', 'shale']
sedimentary.sort()
sedimentary
```

## Deleting Elements
* pop returns the element that was removed. It deletes and returns the last element if no index was provided.
* del
* remove

```
m = ['gneiss', 'marble', 'schist', 'slate']
m1 = m.pop()
print(m)
print(m1)
```

```
m = ['gneiss', 'marble', 'schist', 'slate']
m2 = m.pop(2)
print(m)
print(m2)
```

```
m = ['gneiss', 'marble', 'slate']
del m[1]
m
```

```
m = ['gneiss', 'marble', 'slate', 'schist']
m.remove('slate')
m
```

```
igneous = ['basalt', 'granite', 'breccia', 'pumice', 'diabase']
del igneous[1:3]
igneous
```

## List, dictionary, Tuple, Set

```
a = ['sandstone', 'mudstone', 'basalt'] # Create a new list with some data.
b = {'sandstone', 'mudstone', 'basalt'} # A set is an unordered collection of distinct elements.
c = (5,6,7,8) # A tuple is an immutable ordered list of values.
d = {'sandstone':'sedimentary', 'basalt':'igneous'} # A dictionary stored (Key, value) pairs.
print(a)
print(b)
print(c)
print(d)
```
